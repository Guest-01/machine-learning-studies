# Limitation of Linear Regression
선형 회귀는 직선을 그리기 때문에 직선이 아닌 관계에서는 취약하다는 것을 확인해보자

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 100)
y = X**2
plt.title("OriginalData")
plt.plot(X, y)
plt.show()
```

```python
X = X.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

lr = LinearRegression().fit(X, y)
dt = DecisionTreeRegressor().fit(X, y)
```

```python
from sklearn.metrics import mean_absolute_error

lr_score = mean_absolute_error(y, lr.predict(X))
dt_score = mean_absolute_error(y, dt.predict(X))

print(f"{lr_score=}\n{dt_score=}")
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title("LinearRegression")
plt.plot(X, lr.predict(X))
plt.subplot(1, 2, 2)
plt.title("DecisionTree")
plt.plot(X, dt.predict(X))
plt.show()
```
