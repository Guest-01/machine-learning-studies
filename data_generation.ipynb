# Data Generation
scikit-learn 라이브러리에는 모델 테스트를 위해 데이터셋을 만들어주는 함수들이 있다.   
간단하게 모델을 테스트해보고 싶은데 적당한 데이터셋이 없다면 직접 만들어볼 수 있는 유용한 함수들이다.

여기서 간단히 알아볼 함수는 아래 세개이며 각각의 파라미터는 공식문서를 참고하도록 하자.   
또한 여기 세개 외에도 다양한 함수가 있으니 마찬가지로 공식문서를 확인하자.
- `make_classification` : 분류문제를 위한 데이터셋 만들기
- `make_regression` : 회귀문제를 위한 데이터셋 만들기
- `make_moons` : 반원 모양의 분류 데이터셋 만들기

https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification, make_regression, make_moons
```

```python
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, random_state=1)

print(X.shape, y.shape)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
X, y = make_regression(n_features=2, random_state=1)

print(X.shape, y.shape)

plt.scatter(X[:, 0], y)
plt.scatter(X[:, 1], y)
plt.show()
```

```python
X, y = make_moons(random_state=1)

print(X.shape, y.shape)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```
